```python
import pandas as pd
df.dropna(subset=['Accident_Spot', 'Accident_Location', 'Severity', 'Collision_Type', 'Road_Character', 'Road_Type'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
unique_accident_spots = df['Severity'].unique()
print("Unique values in 'Accident_Spot' column:")
for value in unique_accident_spots:
    print(value)
```

```python
# Handle NaN values (if any) in the DataFrame
df.fillna("Unknown", inplace=True)  # Replace NaN values with "Unknown" or any other appropriate value

# Define mapping dictionaries for each categorical feature
accident_spot_mapping = {'Bottleneck': 0, 'Road hump or Rumble strips': 1, 'Cross roads': 2,'Bridge': 3,'Junction': 4,'Curves': 5,'Culvert': 6,'Narrow road': 7,'T Junction': 8,'Railway crossing': 9,'Offset': 10,'Circle': 11,'Y Junction': 12,'Staggered junction': 13,'Circle': 14,'Y Junction': 15,'Staggered junction': 16,'More than four arms': 17,'Rail Crossing manned': 18,'Round about or Circle': 19,'Rail Crossing Unmanned': 20}
accident_location_mapping = {'Rural Areas': 0, 'City/Town': 1, 'Villages settlement': 2}
collision_type_mapping = {'Drowned': 0, 'Skidding or Self accident': 1, 'Hit bicyclist': 2, 'Head on': 3, 'Rear end': 4, 'Hit puchcart': 5, 'Hit fixed object': 6, 'Overturning': 7, 'Vehicle to Vehicle': 8, 'Hit pedestrian': 9, 'Hit animal': 10, 'Side impact or Right angle': 11, 'Side swipe': 12, 'Run Off Road': 13 , 'Hit and Run': 14, 'Right Turn Collision': 15, 'Hit parked vehicle': 16, 'Hit pedal cyclist': 17}
road_character_mapping = {'Curve': 0, 'Hump': 1, 'Dip or trough': 2 , 'Straight and flat': 3, 'Curve and Incline': 4, 'Incline': 5, 'Crest of hill': 6, 'Slight Curve': 7, 'Sharp Curve': 8, 'Steep Incline or Climb': 9, 'Gentle Incline or Climb': 10}
road_type_mapping = {'State Highway': 0, 'Village Road': 1, 'NH': 2, 'One way': 3, 'City or Town Road': 4, 'Major District Road': 5, 'Two way': 6, 'Sub Arterial': 7, 'Service Road': 8, 'Feeder Road': 9, 'Expressway': 10, 'Arterial': 11, 'Mixed': 12, 'Minor District Road': 13, 'Residential Street': 14, 'Forest Road': 15}
severity_mapping = {'Grievous Injury': 0, 'Fatal': 1, 'Damage Only': 2, 'Simple Injury': 3}

# Define a function to label encode categorical columns
def label_encode_column(df, column_name, mapping):
    df[column_name] = df[column_name].map(mapping)
    return df

# Apply label encoding to each categorical column
df = label_encode_column(df, 'Accident_Spot', accident_spot_mapping)
df = label_encode_column(df, 'Accident_Location', accident_location_mapping)
df = label_encode_column(df, 'Collision_Type', collision_type_mapping)
df = label_encode_column(df, 'Road_Character', road_character_mapping)
df = label_encode_column(df, 'Road_Type', road_type_mapping)
df = label_encode_column(df, 'Severity', severity_mapping)  # Also encode the target variable if needed

# Display the updated DataFrame
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X = df[['Accident_Spot', 'Accident_Location', 'Collision_Type', 'Road_Character', 'Road_Type']]
y = df['Severity']

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the decision tree classifier
clf = DecisionTreeClassifier()

# Train the model on the training data
model=clf.fit(X_train, y_train)

# Predict severity on the testing data
y_pred = clf.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
new_data = {
    'Accident_Spot': 3,  # Example numerical value
    'Accident_Location': 1,  # Example numerical value
    'Collision_Type': 0,  # Example numerical value
    'Road_Character': 2,  # Example numerical value
    'Road_Type': 12  # Example numerical value
}

# Convert the dictionary into a DataFrame
new_data_df = pd.DataFrame([new_data])

# Make predictions on the new data
severity_prediction = clf.predict(new_data_df)

print("Predicted Severity:", severity_prediction)
```

```python
# Example: Predict the severity for new data
new_data = pd.DataFrame({'Accident_Spot': [0], 'Accident_Location': [0], 'Collision_Type': [0], 'Road_Character': [0], 'Road_Type': [0]})
predicted_severity = model.predict(new_data)
print("Predicted Severity:", predicted_severity)
```

```python
# Reverse severity mapping dictionary
reverse_severity_mapping = {v: k for k, v in severity_mapping.items()}

# Convert numerical severity predictions back to categorical values
predicted_severity_categorical = [reverse_severity_mapping[prediction] for prediction in severity_prediction]

print("Predicted Severity (Categorical):", predicted_severity_categorical)
```
